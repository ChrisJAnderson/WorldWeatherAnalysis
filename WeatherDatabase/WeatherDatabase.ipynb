```python
#dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from citipy import citipy
import requests
from config import APIKey
import time
from datetime import datetime
```

```python
#generate lat/long zip

Lat= np.random.uniform(-90.000, 90.000,size=2000)
Long= np.random.uniform(-90.000,90.000,size=2000)
LatLong= zip(Lat,Long)
```

```python
#unpack lat/long zip
Coords= list(LatLong)
```

```python
#generate nearest cities
Cities=[]
#identifies city
for coordinate in Coords: 
    City=(citipy.nearest_city(coordinate[0],coordinate[1]).city_name)
    #and adds unique cities to list
    if City not in Cities:
        Cities.append(City)
```

```python
#url for API Call
url="http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID="+APIKey
# Create an endpoint URL for a city.
city_url = url + "&q=" + "Boston"
city_weather = requests.get(city_url)
city_weather.json()
```

```python
#Define lists and variables for weather data
CityData=[]
SetCount=1
RecordCount=1
#loop cities in Cities to add to CityData
for i, city in enumerate(Cities):
    #run in sets of 50 to prevent timeout errors
    if (i%50==0 and i>=50):
        SetCount+=1
        RecordCount=1
        time.sleep(60)
    #create url for city in Cities
    CityUrl = url + "&q=" + city.replace(" ","+")
    #print city, record, and set numbers and increase record count
    print(f"{RecordCount} of {SetCount}: {city}")
    RecordCount+=1
    try:
        # Parse JSON
        CityWeather = requests.get(CityUrl).json()
        # Parse requested data
        CityLat = CityWeather["coord"]["lat"]
        CityLong = CityWeather["coord"]["lon"]
        CityMaxTemp = CityWeather["main"]["temp_max"]
        CityHumid = CityWeather["main"]["humidity"]
        CityCloud = CityWeather["clouds"]["all"]
        CityWind = CityWeather["wind"]["speed"]
        CityDesc=CityWeather["weather"][0]["description"]
        CityCountry=CityWeather["sys"]["country"]
        print(CityDesc)
        # Append to city in CityData
        CityData.append({"City": city.title(),
                          "Lat": CityLat,
                          "Lng": CityLong,
                          "Max Temp": CityMaxTemp,
                          "Humidity": CityHumid,
                          "Cloudiness": CityCloud,
                          "Wind Speed": CityWind,
                        "Desc":CityDesc,
                        "Country":CityCountry})
# Catch invalid coords
    except:
        print("City not found. Skipping...")
        pass

# finish retrieval
print("Done")
```

```python
CityDataDF=pd.DataFrame(CityData)
CityDataDF
```

```python
CityDataDF.to_csv('WeatherDatabase.csv',index=False)
```
